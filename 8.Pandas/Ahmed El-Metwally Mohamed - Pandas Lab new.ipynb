## Get the lower and higher limit of the salary in new columns (Hint: use regex, split)

```python
new=df['Salary Estimate'].str.split('-',n=1,expand=True)
df["Minimum Salary"]=new[0].str.extract(r'(\d+)').astype(int)
df["Maximum Salary"]=new[1].str.extract(r'(\d+)').astype(int)
```

```python
df.head()
```

## get average salary

```python
# Make a new column 
df['Average Salary'] = (df['Minimum Salary'] + df['Maximum Salary'])/2
df
```

## Extract state code from "location" column as a new column "state"

```python
df['state']=df['Location'].str.extract(r',\s*([A-Za-z]*)')
df
```

## Get values count of 'state' column

```python
df['state'].value_counts()
```

## replace values using the next dict
state_replacements = {

```python
state_replacements = {
    'United States': 'US',
    'Texas': 'TX',
    'California': 'CA',
    'New Jersey': 'NJ',
    'Remote': np.nan,
    'Utah': np.nan 
}

df['state']=df['state'].replace(state_replacements)
df['state'].value_counts()
```

## Drop na from rows and reset index

```python
df=df.dropna().reset_index()
df
```

```python
df.shape
```

### make a new feature for each word in the following list ["python", "excel", "hadoop", "spark", "aws", "tableau", "big data"] as binary  column [0,1] if the word appeared In 'Job Description' column

```python
list=["python", "excel", "hadoop", "spark", "aws", "tableau", "big data"] 
for item in list:
      df[item]=df['Job Description'].str.contains(item,case=False ).astype(int)
```

## Plot mean salary per Sector using the mean of avg salary column you created earlier

```python
import matplotlib.pyplot as plt
pivot_table = pd.pivot_table(df, values='Average Salary', index='Sector', aggfunc=np.mean)
pivot_table.plot(kind='bar')
plt.xlabel('Sector')
plt.legend()
plt.show()
```
